**Contract**

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
  if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
    print(row['Contract'])
```

```python
def extract_contract_info(contract):
  if 'On Loan' in contract or 'Free' in contract:
    start_date = np.nan
    end_date = np.nan
    contract_len = 0
  else:
    start_date, end_date = contract.split(" ~ ")
    start_yr = int(start_date[:4])
    end_yr = int(end_date[:4])
    contract_len = end_yr - start_yr

  return start_date, end_date, contract_len

cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x : pd.Series(extract_contract_info(x)))

for i in range(len(cols)):
  fifa.insert(loc = fifa.columns.get_loc('Contract') + 1 + i, column = cols[i], value = new_data[i])
```

```python
fifa.sample(5)
```

```python

fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
def set_contract_status(contract):
  if 'Free' in contract:
    return 'Free'
  elif 'On Loan' in contract:
    return 'On Loan'
  else:
    return 'Contract'

fifa.insert(loc = fifa.columns.get_loc('Contract Length(years)') + 1, column ='Contract Status', value = fifa['Contract'].apply(set_contract_status))
```

**Height**

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
  if 'cm' in height:
    return int(height.strip('cm'))
  else:
    feet, inches = height.split("'")
    total_inches = int(feet) * 12 + int(inches.strip('"'))
    return round(total_inches * 2.54)

fifa['Height'] = fifa['Height'].apply(convert_height)
```

```python
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns = {'Height' : "Height(cm)"})
```

**Weight**

```python
def convert_weight(weight):
  if "kg" in weight:
    return int(weight.strip("kg"))
  else:
    pounds = int(weight.strip("lbs"))
    return round(pounds / 2.205)


fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

**Hits**

```python
fifa['Hits'].dtype
```

```python
fifa['Hits'].unique()
```

```python
def convert_searches(hit):
  if type(hit) == str:
    if 'K' in hit:
      num = float(hit.strip('K'))
      return int(num * 1000)
    return int(hit)
  if np.isnan(hit):
    return 0
  elif type(hit) == float:
    return int(hit)
  elif type(hit) == int:
    return hit
  return 0

fifa['Hits'] = fifa['Hits'].apply(convert_searches)
```

**Value**

```python
fifa['Value'].dtype
```

```python
def conv_value(val):
  val = val.strip('€')
  if 'M' in val:
    val = float(val.strip('M'))
    return int(val * 1000)
  return int(val.strip('K'))

fifa['Value'] = fifa['Value'].apply(conv_value)
```

**Wage**

```python
fifa['Wage'].unique()
```

```python
def conv_wage(wage):
  wage = wage.strip('€')
  if wage[-1] == 'K':
    return float(wage.strip('K'))
  elif wage == '0':
    return 0.0
  return int(wage)/1000

fifa['Wage'] = fifa['Wage'].apply(conv_wage)
```

```python
fifa['Wage'].unique()
```

```python
fifa = fifa.rename(columns = {'Wage' : "Wage(€ K)"})
fifa.sample(3)
```
